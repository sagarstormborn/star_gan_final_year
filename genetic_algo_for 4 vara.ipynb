```python
import random

class agent:

    def __init__(self,max_value=50, max_value2=100, max_value3=250,max_value4=90):
        length=len(bin(max_value)[2:])
        length2=len(bin(max_value2)[2:])
        length3=len(bin(max_value3)[2:])
        length4=len(bin(max_value4)[2:])
        

        self.value=bin(random.randint(0,max_value))[2:]
        self.value2=bin(random.randint(0,max_value2))[2:]
        self.value3=bin(random.randint(0,max_value3))[2:]
        self.value4=bin(random.randint(0,max_value4))[2:]

        padd=str(0)*(length-len(self.value))
        padd2=str(0)*(length2-len(self.value2))
        padd3=str(0)*(length3-len(self.value3))
        padd4=str(0)*(length4-len(self.value4))

        self.value=padd+self.value
        self.value2=padd2+self.value2
        self.value3=padd3+self.value3
        self.value4=padd4+self.value4


        self.int_value=int(self.value,2)
        self.int_value2=int(self.value2,2)
        self.int_value3=int(self.value2,2)
        self.int_value4=int(self.value4,2)


        self.fitness = -800

        self.max_value=max_value
        self.max_value2=max_value2
        self.max_value3=max_value3
        self.max_value4=max_value4

    def __str__(self):
        return 'value1 : '+str(self.int_value)+"  value2 :"+str(self.int_value2)+"  value3 :"+str(self.int_value3)+"  value4 :"+str(self.int_value4)+' Fitness: ' + str(self.fitness)




def init_agents(population):
  return [agent() for _ in range(population)]
  
def function(x,y,z,q):
  return (+x**2 +3*y**2 +7*x +8*y-8+z**2+7*z-2*q**2+8*q)

def fitness(agents):
  for  a in agents:
    a.fitness=function(a.int_value,a.int_value2,a.int_value3,a.int_value4) 
  # agents = sorted(agents, key=lambda agent: agent.fitness, reverse=True) 
  return agents

def sort_(agents):
  return sorted(agents, key=lambda agent: agent.fitness, reverse=True)

def selection(agents,best=0.2):
  agents = sorted(agents, key=lambda agent: agent.fitness, reverse=True) 
  agents=agents[:int(best*len(agents))]
  return agents 


def crossover(agents):
  offspring = []

  for _ in range((population - len(agents)) // 2):
    parent1 = random.choice(agents)
    parent2 = random.choice(agents)
    child1 = agent()
    child2 = agent()


    min_len=min(len(parent1.value),len(parent1.value))
    split = random.randint(0,min_len)

    min_len2=min(len(parent1.value2),len(parent1.value2))
    split2 = random.randint(0,min_len2)

    min_len3=min(len(parent1.value3),len(parent1.value3))
    split3 = random.randint(0,min_len3)

    min_len4=min(len(parent1.value4),len(parent1.value4))
    split4 = random.randint(0,min_len4)

    #value 1 cross over 
    child1.value = parent1.value[:split] + parent2.value[split:]
    child1.int_value=int(child1.value,2)
    if child1.int_value >= child1.max_value:
      child1 = agent()
    child2.value = parent2.value[:split] + parent1.value[split:]
    child2.int_value=int(child2.value,2)
    if child2.int_value >= child2.max_value:
      child2 = agent()


    #value 2 cross over 
    temp1_value=child1.value
    temp1_int_value=child1.int_value

    temp2_value=child2.value
    temp2_int_value=child2.int_value

    child1.value2 = parent1.value2[:split2] + parent2.value2[split2:]
    child1.int_value2=int(child1.value2,2)
    if child1.int_value2 >= child1.max_value2:
      child1 = agent()
      child1.value=temp1_value
      child1.int_value=temp1_int_value

    child2.value2 = parent2.value2[:split2] + parent1.value2[split2:]
    child2.int_value2=int(child2.value2,2)
    if child2.int_value2 >= child2.max_value2:
      child2 = agent()
      child2.value=temp2_value
      child2.int_value=temp2_int_value


    #value 3 cross over 
    temp1_value2=child1.value2
    temp1_int_value2=child1.int_value2

    temp2_value2=child2.value2
    temp2_int_value2=child2.int_value2

    child1.value3 = parent1.value3[:split3] + parent2.value3[split3:]
    child1.int_value3=int(child1.value3,2)
    if child1.int_value3 >= child1.max_value3:
      child1 = agent()
      child1.value=temp1_value
      child1.int_value=temp1_int_value
      child1.value2=temp1_value2
      child1.int_value2=temp1_int_value2


    child2.value3 = parent2.value3[:split3] + parent1.value3[split3:]
    child2.int_value3=int(child2.value3,2)
    if child2.int_value3 >= child2.max_value3:
      child2 = agent()
      child2.value=temp2_value
      child2.int_value=temp2_int_value
      child2.value2=temp2_value2
      child2.int_value2=temp2_int_value2


    #value 3 cross over 
    temp1_value3=child1.value3
    temp1_int_value3=child1.int_value3

    temp2_value3=child2.value3
    temp2_int_value3=child2.int_value3

    child1.value4 = parent1.value4[:split4] + parent2.value4[split4:]
    child1.int_value4=int(child1.value4,2)
    if child1.int_value4 >= child1.max_value4:
      child1 = agent()
      child1.value=temp1_value
      child1.int_value=temp1_int_value
      child1.value2=temp1_value2
      child1.int_value2=temp1_int_value2
      child1.value3=temp1_value3
      child1.int_value3=temp1_int_value3


    child2.value4 = parent2.value4[:split4] + parent1.value4[split4:]
    child2.int_value4=int(child2.value4,2)
    if child2.int_value4 >= child2.max_value4:
      child2 = agent()
      child2.value=temp2_value
      child2.int_value=temp2_int_value
      child2.value2=temp2_value2
      child2.int_value2=temp2_int_value2
      child2.value3=temp2_value3
      child2.int_value3=temp2_int_value3

    offspring.append(child1)
    offspring.append(child2)

    agents.extend(offspring)
  agents=sort_(agents)
  return agents
  
def mutation(agents,best=0.2,mutation_pro=0.1):
  for agent in agents[int(len(agents)*best):]:

    for idx, param in enumerate(agent.value):

      if random.uniform(0.0, 1.0) <= mutation_pro:

        temp = agent.value[0:idx] + str(random.choice([0,1])) + agent.value[idx+1:]
        val=int(temp,2)
        if agent.max_value > val:
          agent.value=temp
          agent.int_value=val

    for idx, param in enumerate(agent.value2):

      if random.uniform(0.0, 1.0) <= mutation_pro:

        temp2 = agent.value2[0:idx] + str(random.choice([0,1])) + agent.value2[idx+1:]
        val2=int(temp2,2)
        if agent.max_value2 > val2:
          agent.value2=temp2
          agent.int_value2=val2

    for idx, param in enumerate(agent.value3):

      if random.uniform(0.0, 1.0) <= mutation_pro:

        temp3 = agent.value3[0:idx] + str(random.choice([0,1])) + agent.value3[idx+1:]
        val3=int(temp3,2)
        if agent.max_value3 > val3:
          agent.value3=temp3
          agent.int_value3=val3

    for idx, param in enumerate(agent.value4):

      if random.uniform(0.0, 1.0) <= mutation_pro:

        temp4 = agent.value4[0:idx] + str(random.choice([0,1])) + agent.value4[idx+1:]
        val4=int(temp4,2)
        if agent.max_value4 > val4:
          agent.value4=temp4
          agent.int_value4=val4



  agents=sort_(agents)
  return agents
```

```python
population = 80
generations = 5000

import tqdm
agents = init_agents(population)
condition=-70
times=0
for gen in tqdm.tqdm(range(generations)):
  agents=fitness(agents)
  agents=selection(agents)
  agents=crossover(agents)

  # for i in agents[:2]:
  #   print(i)

  agents=mutation(agents)

  # print("generation : ",gen)
  for i in agents[:1]:
    # print(i)
    if i.fitness >condition:
      condition=i.fitness 
      times=0
    elif i.fitness == condition:
      times+=1
    else:
      times=0

    if times>=70:
      print("\ngeneration : ",gen)
      print("condition fullfilled solution is ")
      print(i)
      
      break 
    # print(i)
  else:
        continue
  break
   
```
