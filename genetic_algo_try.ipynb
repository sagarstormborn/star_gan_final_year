```python
import random

class agent:

    def __init__(self,max_value=80):
        length=len(bin(max_value)[2:])
        self.value=bin(random.randint(0,max_value))[2:]
        padd=str(0)*(length-len(self.value))
        self.value=padd+self.value
        self.int_value=int(self.value,2)
        self.fitness = -1
        self.max_value=max_value

    def __str__(self):
        return 'value : '+str(self.int_value)+' Fitness: ' + str(self.fitness)




def init_agents(population):
  return [agent() for _ in range(population)]
  
def function(x):
  return (-x**2 +4*x +9 )

def fitness(agents):
  for  a in agents:
    a.fitness=function(a.int_value)   
  agents = sorted(agents, key=lambda agent: agent.fitness, reverse=True) 
  return agents

def selection(agents,best=0.2):
  agents = sorted(agents, key=lambda agent: agent.fitness, reverse=True) 
  agents=agents[:int(best*len(agents))]
  return agents 


def crossover(agents):
  offspring = []

  for _ in range((population - len(agents)) // 2):
    parent1 = random.choice(agents)
    parent2 = random.choice(agents)
    child1 = agent()
    child2 = agent()
    min_len=min(len(parent1.value),len(parent1.value))
    split = random.randint(0,min_len)
    child1.value = parent1.value[:split] + parent2.value[split:]
    child1.int_value=int(child1.value,2)
    if child1.int_value >= child1.max_value:
      child1 = agent()
    child2.value = parent2.value[:split] + parent1.value[split:]
    child2.int_value=int(child2.value,2)
    if child2.int_value >= child2.max_value:
      child2 = agent()

    offspring.append(child1)
    offspring.append(child2)

    agents.extend(offspring)
  agents=fitness(agents)
  return agents
  
def mutation(agents):
  for agent in agents:

    for idx, param in enumerate(agent.value):

      if random.uniform(0.0, 1.0) <= 0.1:

        temp = agent.value[0:idx] + str(random.choice([0,1])) + agent.value[idx+1:]
        val=int(temp,2)
        if agent.max_value > val:
          agent.value=temp
          agent.int_value=val
  agents=fitness(agents)
  return agents
```

```python
population = 10
generations = 200

agents = init_agents(population)

for gen in range(generations):
  agents=fitness(agents)
  agents=selection(agents)
  agents=crossover(agents)

  # for i in agents[:2]:
  #   print(i)

  agents=mutation(agents)

  print("generation : ",gen)
  for i in agents[:2]:
    print(i)
```
